```python
#load libraries and datasets
import pandas as pd
import matplotlib.pyplot as plt
# Read the 'smaoke.csv' file
data = pd.read_csv('smoking.csv')
# Display the first few rows of the DataFrame
data.head()
```

```python
# Define a dictionary to specify the new column names
new_column_names = {
    'Country': 'Country',
    'Year': 'Year',
    'Data.Daily cigarettes': 'Daily_Cigarettes',
    'Data.Percentage.Male': 'Percentage_Male',
    'Data.Percentage.Female': 'Percentage_Female',
    'Data.Percentage.Total': 'Percentage_Total',
    'Data.Smokers.Total': 'Smokers_Total',
    'Data.Smokers.Female': 'Smokers_Female',
    'Data.Smokers.Male': 'Smokers_Male'
}
# Rename the columns using the rename() method
data.rename(columns=new_column_names, inplace=True)
```

```python
#***ASIAN RECORDS OF MALES SMOKERS ***WHO DATASETS
# ASIANS are the folowing::: COUNTRY, Brunei Darussalam, Cambodia, Indonesia, Laos,Malaysia,Myanmar (Burma),Philippines
# Singapore, Thailand, Vietnam
```

```python
# Create an array containing the selected countries
selected_countries = ["Philippines", "Brunei","Cambodia","Indonesia","Laos","Malaysia","Myanmar","Singapore","Thailand","Vietnam"]

# Filter the data to include only the selected countries and Year==2023
filtered_data = data[(data['Country'].isin(selected_countries)) & (data['Year'] == 2012)]


# Calculate the sum of 'Smokers_Male' for the selected countries
sum_smokers_male = filtered_data.groupby('Country')['Smokers_Male'].sum()
# Sort the countries in descending order based on the sum of 'Smokers_Male'
sum_smokers_male = sum_smokers_male.sort_values(ascending=False)
# Create a pie chart
plt.figure(figsize=(7, 10))
plt.pie(sum_smokers_male, labels=sum_smokers_male.index, autopct='%1.2f%%', startangle=140)
plt.title('Relationship between "ASEAN" Countries and Male Smokers')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create an array containing the selected countries
selected_countries = ["Philippines", "Brunei", "Cambodia", "Indonesia", "Laos", "Malaysia", "Myanmar", "Singapore", "Thailand", "Vietnam"]

# Filter the data to include only the selected countries and Year == 2012
filtered_data = data[(data['Country'].isin(selected_countries)) & (data['Year'] == 2012)]

# Calculate the sum of 'Smokers_Male' for the selected countries
sum_smokers_male = filtered_data.groupby('Country')['Smokers_Male'].sum()
# Sort the countries in descending order based on the sum of 'Smokers_Male'
sum_smokers_male = sum_smokers_male.sort_values(ascending=False)

# Create a bar graph
plt.figure(figsize=(5, 5))
plt.bar(sum_smokers_male.index, sum_smokers_male.values)
plt.xlabel('Country')
plt.ylabel('Total Male Smokers')
plt.title('Relationship between "ASEAN" Countries and Male Smokers in 2012')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Print the ranking information
print("Ranking of Countries by Male Smokers (Descending Order) for Year 2012:")
for rank, (country, smokers) in enumerate(sum_smokers_male.items(), start=1):
    print(f"{rank}. {country}: {smokers} Male Smokers")
```
